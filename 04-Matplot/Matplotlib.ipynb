# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10)) # (width x height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title= "Simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

# Making figures with NumPy arrays

We want: 
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace (0, 10, 100) #100 numbers from 0 to 10
x[:10] #show the first 10 numbers
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 15}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title= "Nut Butter Prices", 
       ylabel = "Price ()");
```

```python
fig, ax = plt.subplots()
#For horizontal bar 'barh' we convert inputs into lists
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title= "Nut butter prices",
       xlabel = "Price ()");
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols= 2,
                                             figsize=(10, 5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
sigma, mu = (3, 50)
ax4.hist(sigma*np.random.randn(1000) + mu);
```
